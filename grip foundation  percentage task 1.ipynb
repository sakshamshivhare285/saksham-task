```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
data=pd.read_excel('/content/drive/MyDrive/percentage.xlsx')
```

```
data.head()
```

```
data.plot()
```

```
sns.countplot(data=data)
```

```
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```
inputs=data[['Hours']]
target=data[['Scores']]
```

```
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest= train_test_split(inputs, target, test_size=0.2, random_state=42)
```

```
model=lr.fit(xtrain,ytrain)
```

```
predictions=model.predict(xtest).round()
```

```
from sklearn.metrics import confusion_matrix, classification_report
```

```
print(ytest)
```

```
print(predictions)
```

```
hours=[[9.25]]
own_pred=lr.predict(hours)
print("no. of hours={}".format(hours))
print(" predicted scores={}".format(own_pred[0]))
```

```
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(ytest,predictions)) 
```
