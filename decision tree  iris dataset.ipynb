```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
data=pd.read_csv('/content/drive/MyDrive/Iris.csv')
```

```
features=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species']
```

```
data.head()
```

```
inputs=data.drop(['Species'], axis=1)
targets=data['Species']
```

```
from sklearn.tree import DecisionTreeClassifier
DT=DecisionTreeClassifier(criterion='gini')
```

```
model=DT.fit(inputs,targets)
```

```
model
```

```
from sklearn.tree import plot_tree
from sklearn import tree
tree.plot_tree(model)
```

```
# Import necessary libraries for graph viz
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

# Visualize the graph
dot_data = StringIO()
export_graphviz(model, out_file=dot_data, feature_names=features,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
